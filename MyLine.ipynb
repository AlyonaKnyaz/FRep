```python
from tkinter import Tk, Canvas, PhotoImage, mainloop
import math

WIDTH, HEIGHT = 640, 480

window = Tk()
canvas = Canvas(window, width=WIDTH, height=HEIGHT, bg="#ffffff")
canvas.pack()
img = PhotoImage(width=WIDTH, height=HEIGHT)
canvas.create_image((WIDTH/2, HEIGHT/2), image=img, state="normal")

x1,y1=200,100
x2,y2=300,500

def MyLine(x1,y1,x2,y2):
    if x2<x1:
        a=x1
        x1=x2
        x2=a
        b=y1
        y1=y2
        y2=b
    dy=y2-y1
    dx=x2-x1
    k=dy
    if dx==0:
        x=x1;
        for y in range(y1,y2):
            img.put("#000000",(x,y))
            
    if dy==0:
        y=y1;
        for x in range(x1,x2):
            img.put("#000000",(x,y))
            
    if math.fabs(dy) >= math.fabs(dx) and k>0:
        e=0
        x=x2
        for y in range(y2,y1,-1):
            e=e+dx
            if e>0.5*dx:
                x=x-1
                e=e-dy
            img.put("#000000",(x,y))
            
    if math.fabs(dy) <= math.fabs(dx) and k>0:
        e=0
        y=y1
        for x in range(x1,x2+1):
            e=e+k
            if e>0.5*dx:
                y=y+1
                e=e-dx
            img.put("#000000",(x,y))
            
    if math.fabs(dy) >= math.fabs(dx) and k<0:
        e=0
        x=x2
        for y in range(y2,y1):
            e=e-k
            if e>0.5*dx:
                x=x-1
                e=e+dy
            img.put("#000000",(x,y))
            
    if math.fabs(dy) <= math.fabs(dx) and k<0:
        e=0
        y=y1
        for x in range(x1,x2+1):
            e=e-k
            if e>0.5*dx:
                y=y-1
                e=e-dx
            img.put("#000000",(x,y))
        
            
MyLine(400,100,200,400)

mainloop()
```
