```python
P=[['S','aA'],['A','aAB'],['A','a'],['B','b'],['B','']]
Vt=['a','b']
Vn=['S','A','B']
S='S'

#Проверка на принадлежность к КС-грамматике
def Type2(P):
    a=1
    for i in P:
        for j in Vt:
            if j in i[0]:
                a=0
                return(False)
        if i[0]=='':
            a=0
            return(False)
    if a==1:
        return(True)
    
def Type3(P):
    a=1
    c=0
    for i in P:
        c=0
        for j in Vt:
            if j in i[0]:
                a=0
                return(False)
        for n in Vt:
            if n in i[1]:
                c=1
        
        if c==0:
            return(False)
        
        for l in Vn:
            if l in i[1] and i[1].find(l) != len(i[1])-1:
                for k in range(i[1].find(l)+1,len(i[1])):
                    for h in Vt:
                        if i[1][k]==h:
                            a=0
                            return(False)
                      
def LangKC(P):                        
    if Type2(P)==True and Type3(P)==False: 
        return(True)
    else:
        return(False)
    
if LangKC(P)==True:
    print('КС-грамматика')
    

def ShopAuto(lenta):
    stek=[]
    count=[]
    for i in Vn:
        count.append(0)
    cnt=0
    print(' ',stek,lenta)
    stek.append(S)
    print('push',S,'',stek,' ',lenta)
    while(lenta!=''):
        i=stek.pop(0)
        print('pop', i,' ',stek,' ',lenta)
        if i in Vn:
            
            for j in P:
                if j[0]==i:
                    count[Vn.index(i)]+=1
                    
            if count[Vn.index(i)]==1:
                for j in P:
                    if j[0]==i:
                        if j[1]=='':
                            if lenta=='':
                                if stek==[]:
                                    return(True)
                                    break
                        else:
                            for k in j[1][::-1]:
                                stek.append(k)
                                print('push', k,' ',stek,' ',lenta)   
                    
            
            if count[Vn.index(i)]>1:
                for j in P:
                    if cnt==0:
                        if j[0]==i:
                            if len(lenta)>=len(j[1]):
                                cnt+=1
                                if j[1]=='':
                                    if lenta=='':
                                        if stek==[]:
                                            return(True)
                                            break
                                else:
                                    for k in j[1][::-1]:
                                        stek.append(k)
                                        print('push', k,' ',stek,' ',lenta)
                cnt=0
                        
                            
                    
            
        if i in Vt:
            print('read', lenta[0],' ',stek,' ',lenta)
            lenta=lenta[1:]
            if lenta=='':
                if stek==[]:
                    return(True)
                    break
        
                    
    if lenta=='' and stek!=[]:
        return(False)
        
    
if ShopAuto(word1)==True:
    print('Слово word1 распознается автоматом')
else:
    print('Слово word1 не распознается автоматом')
    
print()
if ShopAuto(word2)==True:
    print('Слово word2 распознается автоматом')
else:
    print('Слово word2 не распознается автоматом')
    
word1='aa'
word2='abbb'
```
