```python
import math
import numpy as np
from tkinter import Tk, Canvas, PhotoImage, mainloop
WIDTH, HEIGHT = 640, 480
a=200 #сторона куба
Pnts=[[-0.5,0.5,-0.5],[-0.5,-0.5,-0.5],[0.5,-0.5,-0.5],[0.5,0.5,-0.5],[-0.5,0.5,0.5],[-0.5,-0.5,0.5],[0.5,-0.5,0.5],[0.5,0.5,0.5]]
Ribs=[[0,1],[0,3],[0,4],[1,2],[1,5],[2,3],[2,6],[3,7],[4,5],[4,7],[5,6],[6,7]]
Facet=[[0,3,4],[3,7,4],[3,2,7],[2,6,7],[4,7,5],[7,6,5],[1,5,2],[2,5,6],[0,4,1],[1,4,5],[0,1,3],[3,1,2]]
Facet1=[[0,3,4],[3,7,4],[3,7,2],[2,6,7],[4,7,5],[7,5,6],[1,2,5],[2,5,6],[0,1,4],[1,4,5],[0,1,3],[3,1,2]]

xe=[]
ye=[]
ze=[]
x=[]
y=[]
tetta=60*math.pi/180#math.pi/6
fi=30*math.pi/180#math.pi/4
ro=1000#180
V=[[-math.sin(tetta),-math.cos(fi)*math.cos(tetta),-math.sin(fi)*math.cos(tetta),0],
   [math.cos(tetta), -math.cos(fi)*math.sin(tetta),-math.sin(fi)*math.sin(tetta),0],
   [0              , math.sin(fi),                 -math.cos(fi),                0],
   [0              , 0,                             ro,                          1]]
for i in range(8):
    Pnts[i].append(1)
    A=np.dot(Pnts[i],V)
    xe.append(A[0])
    ye.append(A[1])
    ze.append(A[2])
    
for i in range(8):
    x.append(xe[i]*ro/(2*ze[i]))
    y.append(ye[i]*ro/(2*ze[i]))
    
window = Tk()
canvas = Canvas(window, width=WIDTH, height=HEIGHT, bg="#ffffff")
canvas.pack()
img = PhotoImage(width=WIDTH, height=HEIGHT)
canvas.create_image((WIDTH/2, HEIGHT/2), image=img, state="normal")
#простая отрисовка куба
#for j in Ribs:
#    canvas.create_line(a*x[j[0]]+200,a*y[j[0]]+200,a*x[j[1]]+200,a*y[j[1]]+200)
 
#лицевые грани, способ 1
def Way1():
    count=0
    for l in Facet:
        count+=1
        x1=xe[l[0]]
        x2=xe[l[1]]
        x3=xe[l[2]]
        y1=ye[l[0]]
        y2=ye[l[1]]
        y3=ye[l[2]]
        z1=ze[l[0]]
        z2=ze[l[1]]
        z3=ze[l[2]]
        h=x1*y2*z3-x1*y3*z2+x3*y1*z2-x2*y1*z2+x2*y3*z1-x3*y2*z1
        print(h)
        if h<0:
            if count%2==1:
                xr=x[l[0]]
                yr=y[l[0]]
            if count%2==0:
                #canvas.create_polygon(a*xr+200,a*yr+200,a*x[l[0]]+200,a*y[l[0]]+200,a*x[l[1]]+200,a*y[l[1]]+200,a*x[l[2]]+200,a*y[l[2]]+200,outline='black',fill='green')
                #удаление лишних линий
                canvas.create_line(a*xr+200,a*yr+200,a*x[l[0]]+200,a*y[l[0]]+200,a*x[l[1]]+200,a*y[l[1]]+200,a*x[l[2]]+200,a*y[l[2]]+200,a*xr+200,a*yr+200)

def Way2():
    count=0
    xc=0
    yc=0
    zc=0
    for i in range(8):
        xc+=xe[i]
        yc+=ye[i]
        zc+=ze[i]
    xc=xc/8
    yc=yc/8
    zc=zc/8
    for l in Facet1:
        count+=1
        x1=xe[l[0]]
        x2=xe[l[1]]
        x3=xe[l[2]]
        y1=ye[l[0]]
        y2=ye[l[1]]
        y3=ye[l[2]]
        z1=ze[l[0]]
        z2=ze[l[1]]
        z3=ze[l[2]]
        a=y2*z3-y1*z3-y2*z1-y3*z2+y3*z1+y1*z2
        b=x3*z2-x3*z1-x1*z2-x2*z3+x1*z3+x2*z1
        c=x2*y3-x1*y3-x2*y1-x3*y2+x3*y1+x1*y2
        h=x1*y2*z3-x1*y3*z2+x3*y1*z2-x2*y1*z2+x2*y3*z1-x3*y2*z1
        if (a*xc+b*yc+c*zc-h)>=0:
            ls=1
        else:
            ls=-1
        r=math.pow(math.pow(a,2)+math.pow(b,2)+math.pow(c,2),0.5)
        ec=[(2/3)*(x1+x2+x3),(2/3)*(y1+y2+y3),(2/3)*(z1+z2+z3)]
        ecr=math.pow(math.pow(ec[0],2)+math.pow(ec[1],2)+math.pow(ec[2],2),0.5)
        cosa=(ec[0]*a*ls+ec[1]*b*ls+ec[2]*c*ls)/(r*ecr)
        print(cosa,count)
        if cosa>=0 and cosa<=1:
            a=200
            canvas.create_polygon(a*x[l[0]]+200,a*y[l[0]]+200,a*x[l[1]]+200,a*y[l[1]]+200,a*x[l[2]]+200,a*y[l[2]]+200,fill='green')
            
Way2()
    
mainloop()
```
