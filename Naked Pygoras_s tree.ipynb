```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import math
import matplotlib.lines as lines

x=0.5
y=0.5
side=0.1 
deep=12 
alfa=math.pi


fig = plt.figure()
ax = fig.add_axes([0,0,10,10])

def Pyth_tree( x, y, xl, yl, side, fi, alfa, deep):
    dx=side*math.sin(fi)
    dy=side*math.cos(fi)
    x1=x+dx
    y1=y+dy   
    x2=x+dx-dy
    y2=y+dy+dx
    x3=x-dy
    y3=y+dx
    x4=x-dy+side*math.cos(alfa)*math.sin(fi-alfa)
    y4=y+dx+side*math.cos(alfa)*math.cos(fi-alfa)
    
    x5=(x+x1+x2+x3)/4
    y5=(y+y1+y2+y3)/4
    x6=(x2+x3+x4)/3
    y6=(y2+y3+y4)/3
    p1=patches.Polygon([[xl, yl], [x5, y5]], fill=False)
    p2=patches.Polygon([[x5, y5], [x6, y6]], fill=False)
    ax.add_patch(p1)
    ax.add_patch(p2)
    if deep>1:
        Pyth_tree(x4, y4, x6, y6, side*math.sin(alfa), fi-alfa+math.pi/2, alfa, deep-1)
        Pyth_tree(x3, y3, x6, y6, side*math.cos(alfa), fi-alfa, alfa, deep-1)

Pyth_tree(x, y, x+side/2, y, side, math.pi/2, alfa/6, deep) 
ax.set_axis_off() 
plt.show()
```
