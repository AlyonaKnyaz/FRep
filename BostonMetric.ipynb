```python
import pandas
from sklearn import neighbors, model_selection,preprocessing,datasets
Boston=datasets.load_boston()
Boston.data=preprocessing.scale(Boston.data)
```

```python
import numpy
clf=[]
count=0
for p in numpy.linspace(1, 10, num=200):
    if count==0:
        print(p)
    clf.append(neighbors.KNeighborsRegressor(n_neighbors=5,weights='distance', metric='minkowski',p=p))
    count+=1
    
```

```python
kf = model_selection.KFold(n_splits=5,shuffle=True,random_state=42)
score=[]
for i in clf:
    score.append(model_selection.cross_val_score(i, Boston.data, Boston.target, cv=kf, scoring='neg_mean_squared_error'))
```

```python
scoreacc=[]
for i in score:
    scoreacc.append(i.mean())
scoreacc
```

```python
maxi=-100
for i in range(len(scoreacc)):
    if scoreacc[i]>maxi:
        maxi=scoreacc[i]
        j=i
maxi,j
```
