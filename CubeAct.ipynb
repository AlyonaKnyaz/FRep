```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import math
import time
import matplotlib.lines as lines
fig = plt.figure()
plt.axis([0,1000,0,1])
plt.ion()


a=5 #сторона куба
Pnts=[[-0.5,0.5,-0.5],[-0.5,-0.5,-0.5],[0.5,-0.5,-0.5],[0.5,0.5,-0.5],[-0.5,0.5,0.5],[-0.5,-0.5,0.5],[0.5,-0.5,0.5],[0.5,0.5,0.5]]
Ribs=[[0,1],[0,3],[0,4],[1,2],[1,5],[2,3],[2,6],[3,7],[4,5],[4,7],[5,6],[6,7]]

for c in range(0,30):
    xe=[]
    ye=[]
    ze=[]
    x=[]
    y=[]
    tetta=math.pi/4+c*0.3
    fi=math.pi/4
    ro=130+c*5
    V=[[-math.sin(tetta),-math.cos(fi)*math.cos(tetta),-math.sin(fi)*math.cos(tetta),0],
       [math.cos(tetta), -math.cos(fi)*math.sin(tetta),-math.sin(fi)*math.sin(tetta),0],
       [0              , math.sin(fi),                 -math.cos(fi),                0],
       [0              , 0,                             ro,                          1]]
    for i in range(8):
        if c==0:
            Pnts[i].append(1)
        A=np.dot(Pnts[i],V)
        xe.append(A[0])
        ye.append(A[1])
        ze.append(A[2])
        
    for i in range(8):
        x.append(xe[i]*ro/(2*ze[i]))
        y.append(ye[i]*ro/(2*ze[i]))
        
    for j in Ribs:
        plt.plot([a*x[j[0]]+5, a*x[j[1]]+5], [a*y[j[0]]+5, a*y[j[1]]+5])

    plt.show()
    plt.pause(0.001)

   
```
