```python
import numpy as np
import math
from tkinter import Tk, Canvas, PhotoImage, mainloop
WIDTH, HEIGHT = 640, 480
a=200 
Pnts=[]
#Ribs=[[0,1],[0,3],[0,4],[1,2],[1,5],[2,3],[2,6],[3,7],[4,5],[4,7],[5,6],[6,7]]
Ribs=[]
rs=1
k=0
zet=0
cnt=0
for t in range(0,360,5):
    for f in range(0,360,5):
        if k==0 or (k>0 and [rs*math.sin(t*math.pi/180)*math.cos(f*math.pi/180),rs*math.sin(t*math.pi/180)*math.sin(f*math.pi/180),rs*math.cos(t*math.pi/180)] != Pnts[k-1]): 
            Pnts.append([])
            Pnts[k].append(rs*math.sin(t*math.pi/180)*math.cos(f*math.pi/180))
            Pnts[k].append(rs*math.sin(t*math.pi/180)*math.sin(f*math.pi/180))
            Pnts[k].append(rs*math.cos(t*math.pi/180))
            if zet==rs*math.cos(t*math.pi/180) and k>0:
            #print(k)
            #if k!=30:
                Ribs.append([])
                Ribs[cnt].append(k)
                Ribs[cnt].append(k-1)
                cnt+=1
            k+=1
            zet=rs*math.cos(t*math.pi/180)
        #Ribs[k].append((ao)
        #Ribs[k].append(ao+1)
        #k+=1
                       
print(Pnts)
Ribs=Ribs[:-1]

                


xe=[]
ye=[]
ze=[]
x=[]
y=[]
tetta=math.pi/12
fi=math.pi/12
ro=180
V=[[-math.sin(tetta),-math.cos(fi)*math.cos(tetta),-math.sin(fi)*math.cos(tetta),0],
   [math.cos(tetta), -math.cos(fi)*math.sin(tetta),-math.sin(fi)*math.sin(tetta),0],
   [0              , math.sin(fi),                 -math.cos(fi),                0],
   [0              , 0,                             ro,                          1]]
for i in range(k):
    Pnts[i].append(1)
    A=np.dot(Pnts[i],V)
    xe.append(A[0])
    ye.append(A[1])
    ze.append(A[2])
    
for i in range(k):
    x.append(xe[i]*ro/(2*ze[i]))
    y.append(ye[i]*ro/(2*ze[i]))
    
window = Tk()
canvas = Canvas(window, width=WIDTH, height=HEIGHT, bg="#ffffff")
canvas.pack()
img = PhotoImage(width=WIDTH, height=HEIGHT)
canvas.create_image((WIDTH/2, HEIGHT/2), image=img, state="normal")

for j in range(0,k):
    canvas.create_line(a*x[j]+200,a*y[j]+200,a*x[j]+201,a*y[j]+201)

print(Ribs)
    
for n in Ribs:
    canvas.create_line(a*x[n[0]]+200,a*y[n[0]]+200, a*x[n[1]]+200,a*y[n[1]]+200)
    
mainloop()
```
