```python
data['FirstName'] = pandas.Series('', index=data.index)
```

```python
for index, row in data.iterrows():
    if '(' in row['Name'] and '("' not in row['Name']:
        ind=row['Name'].find('(')
        st=row['Name'][(ind+1):]
    else:
        ind=row['Name'].find('.')
        st=row['Name'][(ind+2):]
    ind=st.find(' ')
    #df.loc[row_indexer, column_indexer]
    data['FirstName'][index]=st[:ind]
    #row.loc['FirstName',index]=st[:ind]
    row['FirstName']=st[:ind]
    print(row['FirstName'])
```

```python
data1=data[data['Sex']=='male']
data2=data[data['Sex']=='female']
Daily1 = data1.reset_index().groupby(['FirstName']).count()
Daily2 = data2.reset_index().groupby(['FirstName']).count()
```

```python
print(Daily1['Name'][Daily1['PassengerId']==Daily1['PassengerId'].max()])
print(Daily2['Name'][Daily2['PassengerId']==Daily2['PassengerId'].max()])
```
