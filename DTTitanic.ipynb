```python
import pandas
import numpy as np
from sklearn.tree import DecisionTreeClassifier
data = pandas.read_csv('titanic.csv', index_col='PassengerId')
for index, row in data.iterrows():
    if row['Sex']=='male':
        data['Sex'][index]=0
    else:
        data['Sex'][index]=1
data=data.dropna()
data[100:150]
```

```python
y=[]
for index, row in data.iterrows():
    y.append(row['Survived'])
```

```python
X=[]
count=0
for index, row in data.iterrows():
    X.append([])
    X[count].append(row['Parch'])
    X[count].append(row['Fare'])
    X[count].append(row['Age'])
    X[count].append(row['Sex'])
    count+=1
X
```

```python
clf = DecisionTreeClassifier(random_state=241)
clf.fit(X, y)
```

```python
importances = clf.feature_importances_
```
