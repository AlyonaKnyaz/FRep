```python
import math
import scipy.special as ss
N=52
a=0
b=math.pi
l=[a]

for i in range(1,N+1):
    l.append(l[i-1]+(b-a)/N)
print(l)
```

```python
zn=[]
for i in l:
    e=math.pow((math.sin(i*math.cos(2*i)))/(math.pow(i,2)+5),2)
    zn.append(e)
print(zn)
```

```python
def Pnx(x,sx,sy,n):
    D=1
    P=sy[0]
    sy1=[]
    for k in range(0,len(sy)):
        sy1.append(sy[k])
    for j in range(1,n+1):
        for i in range(0,n-j+1):
            sy1[i]=(sy1[i+1]-sy1[i])/(sx[i+j]-sx[i])
        D=D*(x-sx[j-1])
        P=P+sy1[0]*D
    return(P)      
```

```python
def L(x,num,list1,list2):
    Lag=0
    L1=1
    for i in range(0,num+1):
        for j in range(0,num+1):
            if i != j:
                L1 *= x-list1[j]
                L1 /= list1[i] - list1[j]
        Lag += list2[i]*L1
        L1=1
    return Lag  
```

```python
for p in range(0,N+1):
    print("%.10f" %(l[p]),' ',"%.10f" %zn[p],' ',"%.10f" %L(l[p],N,l,zn),' ',"%.10f" %Pnx(l[p],l,zn,N))
```
