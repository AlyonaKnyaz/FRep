```python
import math
import matplotlib.patches as patches
import matplotlib.pyplot as plt

fig = plt.figure()

def Man(c):
    z=[complex(0,0)]
    a=0
    for i in range(0,20):
        z.append(pow(z[i-1],2)+c)
        if abs(z[i])>2:
            a=1
            return(False)
    if a==0:
        return(True)

clist=[]
i=float(-2)
j=float(-2)
for i in range(-200,200):
    for j in range(-200,200):
        c=complex(i*0.01,j*0.01)
        if Man(c):
            clist.append(c)

    
for i in clist: 
    plt.scatter(i.real,i.imag)
    
    
plt.show()    
```
