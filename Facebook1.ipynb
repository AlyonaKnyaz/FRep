Для вычисления самого популярного объекта выборки, мне кажется, является логичным вычислить нормализованную сумму признаков, 
начиная с Lifetime Post Total Reach и до конца, так как эти признаки оценивают популярность конкретного объекта среди 
пользователей. Для обработки пропусков я заполняю их средним значение по этому признакому, такой подход позволяет смоделировать
реальную ситуацию и найти наиболее популярный объект. Однако, чтобы удостовериться в правильности выбора объекта, я пробую ниже
еще два варианта заполнения пропусков(максимальными и минимальными значениям). Так же, признаки очень разняться по масштабу, чтобу это не влияло на результат, я нормализую признаки(теперь они все варьируются от 0 до 1).
По каждому из способов 120 объект(начиная с 0) является наиболее популярными.

```python
from sklearn import preprocessing 

data=data.fillna(data.mean(numeric_only=True))

X=data.loc[:,'Lifetime Post Total Reach':'Total Interactions'].values

normalized_X = preprocessing.normalize(X)

popularity=[]
count=0
for i in normalized_X:
    popularity.append(0)
    for j in i:
        popularity[count]+=j
    count+=1
```

```python
print(popularity.index(max(popularity)),' ',max(popularity))
```
