```python
import math
N=5
a=0
b=math.pi
l=[a]

for i in range(1,N+1):
    l.append(l[i-1]+(b-a)/N)
print(l)
```

```python
zn=[]
for i in l:
    e=math.pow((math.sin(i*math.cos(2*i)))/(math.pow(i,2)+5),2)
    zn.append(e)
print(zn)
```

```python

def L(x,num,list1,list2):
    Lag=0
    L1=1
    for i in range(0,num+1):
        for j in range(0,num+1):
            if i != j:
                L1 *= x-list1[j]
                L1 /= list1[i] - list1[j]
        Lag += list2[i]*L1
        L1=1
    return Lag    
```

```python
M=20
a=0
b=3.14
l1=[a]

for i in range(1,M+1):
    l1.append(l1[i-1]+(b-a)/M)
print(l1)
```

```python
for i in range(0,M+1):
    print(l1[i]," ",zn1[i]," ",L(l1[i],M,l1,zn1))
```
