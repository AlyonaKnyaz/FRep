```python
Vt=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
Vn=['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O','P','Q','R','S','O','U','V','W','X','Y','Z']

P0=[['S','BCS'],['BCB','D'],['BC','Bc'],['DC','a'],['cS','c'],['aS','a']]
P2=[['S','AS'],['S',''],['A','aA']]
P3=[['S','aS'],['S','bS'],['S','bA'],['A','a']]


def Type0(P):
    for i in P:
        a=0
        for j in Vn:         
            if j in i[0]:
                a=1
        if a==0:
            return(False)
            break     
    if a==1:
        return(True)
            
def Type1(P):
    a=0
    for i in P:
        a=0
        for j in Vn:         
            if j in i[0]:
                a=1
        if a==0:
            return(False)
            break     

        if i[1]=='' and i[0]!='S':
            return(False)
            break
            
    if P.count(['S',''])==0:
        return(False)
       
    if a==1:
        return(True)
    
def Type2(P):
    a=1
    for i in P:
        for j in Vt:
            if j in i[0]:
                a=0
                return(False)
        if i[0]=='':
            a=0
            return(False)
    if a==1:
        return(True)
    
def Type3(P):
    a=1
    c=0
    for i in P:
        c=0
        for j in Vt:
            if j in i[0]:
                a=0
                return(False)
        for n in Vt:
            if n in i[1]:
                c=1
        
        if c==0:
            return(False)
        
        for l in Vn:
            if l in i[1] and i[1].find(l) != len(i[1])-1:
                for k in range(i[1].find(l)+1,len(i[1])):
                    for h in Vt:
                        if i[1][k]==h:
                            a=0
                            return(False)
    if a==1 and c==1:
        return(True)
    
def Typedef(P):
    if Type0(P)==True and Type1(P)==False and Type2(P)==False and Type3(P)==False:
        print(0)
    if Type1(P)==True and Type2(P)==False and Type3(P)==False:
        print(1)
    if Type2(P)==True and Type3(P)==False:
        print(2)
    if Type3(P)==True:
        print(3)
        
Typedef(P0)
Typedef(P2)
Typedef(P3)
```
